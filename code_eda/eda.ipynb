## 3. ***Data Wrangling***

### Data Wrangling Code

	1.	Handling Missing Values
	2.	Removing Outliers
	3.	Dropping Unnecessary Columns

````python
# Proper way to fill missing values
df = df.assign(reviews_per_month=df['reviews_per_month'].fillna(0),
               host_name=df['host_name'].fillna("Unknown"))

# Confirm missing values are fixed
print("Missing values after cleaning:\n", df.isnull().sum())
````

	•	Replace missing reviews_per_month with 0.
	•	Replace missing host_name with "Unknown".
	•	Drop last_review (since it’s not needed).

````python
# Remove listings with price = 0 (not valid)
df = df[df['price'] > 0]

# Remove extremely high prices (cap at $1000)
df = df[df['price'] < 1000]

# Remove unrealistic 'minimum_nights' (cap at 365 days)
df = df[df['minimum_nights'] <= 365]

# Confirm dataset size after cleaning
print("Dataset size after removing outliers:", df.shape)
````

	•	Remove listings with price = 0.
	•	Remove extreme prices above $1000.
	•	Remove minimum_nights > 365 (since Airbnb is usually short-term stays).

````python
# Drop unnecessary columns (if not needed)
df.drop(columns=['id', 'host_id', 'host_name'], inplace=True)

# Confirm new dataset shape
print("New dataset shape:", df.shape)
````


	•	Drop id (just a unique identifier, not useful for analysis).
	
    •	Drop host_id & host_name (if host identity is not needed).
